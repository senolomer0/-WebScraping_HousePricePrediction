# Label Encoder

Machine learning algorithms work better with numerical values. Therefore, we need to convert the categorical variables we have to numerical variables for performance. One way to do this is by using a Label Encoder. 

Label Encoder enumerates categorical variables, starting from 0, according to how many unique values   there are.

```python
le = LabelEncoder()
data.il = le.fit_transform(data.il)
data.ilce = le.fit_transform(data.ilce)
data.emlak_tipi = le.fit_transform(data.emlak_tipi)
data.bulundugu_kat = le.fit_transform(data.bulundugu_kat)
data.isitma = le.fit_transform(data.isitma)
data.balkon = le.fit_transform(data.balkon)
data.esyali = le.fit_transform(data.esyali)
data.kullanim_durumu = le.fit_transform(data.kullanim_durumu)
data.site_icerisinde = le.fit_transform(data.site_icerisinde)
data.krediye_uygun = le.fit_transform(data.krediye_uygun)
data.kimden = le.fit_transform(data.kimden)
data.takas = le.fit_transform(data.takas)
```

```python
data.head()
```

## Models

```python
linear_model = LinearRegression()

knn_model = KNeighborsRegressor()

svr_model = SVR("linear")

dectree_model = DecisionTreeRegressor()

bag_model = BaggingRegressor(bootstrap_features=True)

rf_model = RandomForestRegressor(random_state=42)

grad_model = GradientBoostingRegressor()

xgb_model = XGBRegressor()

light_model = LGBMRegressor()

cat_model = CatBoostRegressor()
```

```python
models = [linear_model, knn_model, svr_model, dectree_model, bag_model, rf_model, grad_model, xgb_model,light_model,cat_model]

model_names = ["LinearRegression","KNeighborsRegressor","SVR","DecisionTreeRegressor","BaggingRegressor",
               "RandomForestRegressor","GradientBoostingRegressor","XGBRegressor","LGBMRegressor","CatBoostRegressor"]
```

Before exporting the data set to the model, we gave all variables except the “fiyat” variable to the x variable and the “fiyat” variable to the y variable. Then we split the data into train and test data. x_train and y_train show the dependent and independent variables to be used to test the model, while x_test and y_test are used to develop the model. Test_size specifies how many % of the data (10%) will be used for testing. Random_state is used to see the same distinction every time we run the program.

```python
x = data.drop("fiyat",axis=1)
y= data.fiyat
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=42)
```

```python
print("TRAIN SCORE\n\n")
print("-----------------------------------------")
for i in range(len(models)):
    model = models[i].fit(x_train,y_train)
    print(model_names[i],"Score = %",r2_score(y_train,models[i].predict(x_train)))
```

```python
print("TEST SCORE\n\n")
print("-----------------------------------------")
for i in range(len(models)):
    model = models[i].fit(x_train,y_train)
    print(model_names[i],"Score = %",r2_score(y_test,models[i].predict(x_test)))
```

According to the results shown above, we observe that Random Forest Regressor, XGBoost Regressor, CatBoost Regressor models give the best results. Therefore, we will try to develop the model over three models.

## XGB Tuned

```python
xgb_params = {"learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
              "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
              "min_child_weight" : [ 1, 3, 5, 7 ],
              "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
              "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]}
```

Repeats Stratified K-Fold repeats n times with different randomization in each repetition.

Randomized search on hyper parameters. The parameters of the estimator used to apply these methods are optimized by cross-validated search over parameter settings.

In contrast to GridSearchCV, not all parameter values are tried out, but rather a fixed number of parameter settings is sampled from the specified distributions. The number of parameter settings that are tried is given by n_iter.

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=10, random_state=1)
random_search = RandomizedSearchCV(xgb_model, xgb_params, cv=cv, random_state=1, n_jobs=-1, verbose=2 ).fit(x_train,y_train)
random_search.best_params_
```

```python
xgb_tuned = XGBRegressor(**random_search.best_params_).fit(x_train,y_train)
y_pred = xgb_tuned.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

# Get Dummies

Categorical variables can transform with the "get_dummies" function in pandas.

After the categorical variables are converted, any transformed new variable belonging to each variable is selected and deleted. Because the value of the deleted variable can be understood by looking at the other transformed variables already remaining.

For example, if all non-deleted variables are 0, it means that the deleted variable must be 1.

```python
df = pd.read_csv("Datasets/data_analyzed.csv")
df.head(3)
```

```python
dms = pd.get_dummies(df[["il","ilce","emlak_tipi","bulundugu_kat","isitma","balkon","esyali","kullanim_durumu","site_icerisinde","krediye_uygun",
                         "kimden","takas"]],drop_first=True)
df = pd.concat([df,dms],axis=1)
df.drop(["il","ilce","emlak_tipi","bulundugu_kat","isitma","balkon","esyali","kullanim_durumu","site_icerisinde","krediye_uygun",
                         "kimden","takas"],axis=1,inplace=True)
df.head(3)
```

```python
x = df.drop("fiyat",axis=1)
y= df.fiyat
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=42)
```

```python
rf_model = RandomForestRegressor().fit(x_train,y_train)
y_pred = rf_model.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

```python
xgb_model = XGBRegressor().fit(x_train,y_train)
y_pred = xgb_model.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

```python
catb_model = CatBoostRegressor().fit(x_train,y_train)
y_pred = catb_model.predict(x_test)
```

```python
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```
