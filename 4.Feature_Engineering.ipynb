There are values in the form of "2+1" in the room variable. This value means 2 bedrooms and 1 living room. Below, we divide the room variable into two as the number of living rooms and the number of rooms and add them to the data set as different variables.

```python
living_room = []
room = []
for i in range(df.shape[0]):
    room.append(df.oda.values[i].split("+")[0])
    living_room.append(df.oda.values[i].split("+")[1])
```

```python
df["salon"] = living_room
df["oda"] = room
```

```python
df.info()
```

```python
df.oda =df.oda.astype("int64")
df.salon = df.salon.astype("int64")
```
